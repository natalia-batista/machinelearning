Implementation of logist regression using maximum likelihood estimation

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from scipy.special import expit
from sklearn.metrics import accuracy_score 
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=23)
```

```python
m = X_train.shape[1] # features
n = X_train.shape[0] # training examples

theta = np.zeros((1,m)) 
        
```

```python
step = 0.001

for it in range(1000):
    gradient = np.zeros(m)

    for j in range(m):
        gradient[j] +=  ((y_train - expit(theta @ X_train.T)) @ X_train[:,j]).item()
        
    theta += step * gradient
```

```python
y_pred = expit(theta @ X_train.T)

# using sklearn accuracy
acc = accuracy_score(y_train,y_pred.T)
print(acc)
```

```python
y_test_pred = expit(theta @ X_test.T)
acc_test = accuracy_score(y_test,y_test_pred.T)
print(acc_test)
```
